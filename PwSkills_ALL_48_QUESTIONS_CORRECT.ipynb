### Q1. What is Unsupervised Learning in the context of Machine Learning?
Unsupervised learning is a type of machine learning where the model is trained on **unlabeled data**.
The algorithm does not know the correct output and must discover patterns, structures, or relationships on its own.

It is mainly used for:
- Clustering
- Dimensionality reduction
- Anomaly detection

Examples include K-Means, DBSCAN, and Hierarchical Clustering.

---

### Q2. How does the K-Means clustering algorithm work?
K-Means is a centroid-based clustering algorithm that divides data into K clusters.

Steps:
1. Choose K
2. Initialize centroids
3. Assign points to nearest centroid
4. Update centroids
5. Repeat until convergence

Objective: minimize intra-cluster variance.

---

### Q3. Explain the concept of a dendrogram in hierarchical clustering
A dendrogram is a tree-like diagram showing how clusters merge at different distance levels.
Cutting the dendrogram gives the final clusters.

---

### Q4. Difference between K-Means and Hierarchical Clustering
K-Means requires predefined clusters and is faster.
Hierarchical clustering builds a tree structure without predefined clusters.

---

### Q5. Advantages of DBSCAN over K-Means
- Detects noise
- No need to specify number of clusters
- Works with arbitrary cluster shapes

---

### Q6. When would you use Silhouette Score?
Used to evaluate clustering quality and select optimal K.

---

### Q7. Limitations of Hierarchical Clustering
- Computationally expensive
- Sensitive to noise
- Not scalable

---

### Q8. Why is feature scaling important in K-Means?
Because K-Means is distance-based and unscaled features bias clustering.

---

### Q9. How does DBSCAN identify noise points?
Points with insufficient neighbors within eps radius are labeled noise.

---

### Q10. Define inertia in the context of K-Means
Inertia is the sum of squared distances of samples to their cluster centroids.

---

### Q11. What is the elbow method in K-Means clustering?
A graphical method to find optimal K where inertia reduction slows.

---

### Q12. Describe the concept of density in DBSCAN
Density refers to the number of points within a neighborhood.

---

### Q13. Can hierarchical clustering be used on categorical data?
Yes, using distance measures like Hamming distance.

---

### Q14. What does a negative Silhouette Score indicate?
Poor clustering where points are closer to other clusters.

---

### Q15. Explain linkage criteria in hierarchical clustering
Defines how distance between clusters is calculated.

---

### Q16. Why K-Means performs poorly on varying cluster sizes or densities?
Because it assumes spherical and equal-sized clusters.

---

### Q17. Core parameters in DBSCAN and their influence
eps and min_samples control cluster density.

---

### Q18. How does K-Means++ improve initialization?
By choosing well-separated initial centroids.

---

### Q19. What is agglomerative clustering?
A bottom-up hierarchical clustering approach.

---

### Q20. Why is Silhouette Score better than inertia?
Because it measures both cohesion and separation.


## PART B â€“ PRACTICAL QUESTIONS (Q21â€“Q48)

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.datasets import load_iris, load_wine, load_breast_cancer, load_digits

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score, silhouette_samples

from scipy.cluster.hierarchy import dendrogram, linkage
```

### Q21. make_blobs (4 centers) + KMeans

```python

X,_=make_blobs(n_samples=300,centers=4,random_state=42)
labels=KMeans(n_clusters=4).fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=labels); plt.show()
```

### Q22. Iris + Agglomerative (3 clusters)

```python

iris=load_iris()
labels=AgglomerativeClustering(n_clusters=3).fit_predict(iris.data)
print(labels[:10])
```

### Q23. make_moons + DBSCAN (outliers)

```python

X,_=make_moons(n_samples=300,noise=0.1)
labels=DBSCAN(eps=0.3).fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=labels); plt.show()
```

### Q24. Wine + StandardScaler + KMeans

```python

wine=load_wine()
X=StandardScaler().fit_transform(wine.data)
labels=KMeans(n_clusters=3).fit_predict(X)
print(np.unique(labels,return_counts=True))
```
