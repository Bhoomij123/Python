## Question 1: Compare and contrast NLTK and spaCy


NLTK is mainly used for research and education and provides many linguistic resources.
spaCy is designed for industrial NLP applications and focuses on performance and speed.

| Feature | NLTK | spaCy |
|---|---|---|
| Purpose | Research/Education | Production |
| Speed | Slower | Faster |

| Models | Few pretrained | Many pretrained |


## Question 2: What is TextBlob?


TextBlob is a Python NLP library built on top of NLTK that simplifies tasks like
sentiment analysis, POS tagging, translation, and noun phrase extraction.

Sentiment analysis outputs:
- **Polarity** (-1 to 1)
- **Subjectivity** (0 to 1)


## Question 3: Role of Stanford NLP


Stanford NLP provides tools for tokenization, POS tagging, named entity recognition,
dependency parsing, and sentiment analysis. It is widely used in academic research
and industrial NLP systems such as chatbots, search engines, and text analytics.


## Question 4: Recurrent Neural Network (RNN)


RNNs are neural networks designed for sequential data. They maintain a hidden state
that allows information from previous inputs to influence future predictions.
They are widely used for language modeling, translation, and speech recognition.


## Question 5: Difference between LSTM and GRU


LSTM has three gates: input, forget, and output gate.
GRU has two gates: update and reset gate.

GRU is simpler and faster, while LSTM handles longer dependencies better.


## Question 6: Sentiment Analysis using TextBlob

```python
from textblob import TextBlob

text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating"""

blob = TextBlob(text)

print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)
```

## Question 7: Tokenization and Frequency Distribution using NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language."""

tokens = word_tokenize(text)

print("Tokens:")
print(tokens)

freq = FreqDist(tokens)

print("\nFrequency Distribution:")
print(freq.most_common(10))
```

## Question 8: LSTM Text Classification using Keras

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np

texts = [
"I love this project",
"This is an amazing experience",
"I hate waiting in line",
"This is the worst service",
"Absolutely fantastic!"
]

labels = [1,1,0,0,1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

seq = tokenizer.texts_to_sequences(texts)
padded = pad_sequences(seq, padding='post')

X = np.array(padded)
y = np.array(labels)

model = Sequential()
model.add(Embedding(100,16,input_length=X.shape[1]))
model.add(LSTM(32))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X,y,epochs=10)
```

## Question 9: spaCy NLP Pipeline

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhabha was an Indian nuclear physicist who played a key role
in the development of India's atomic energy program."""

doc = nlp(text)

for token in doc:
    print(token.text, token.lemma_)

print("\nEntities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
