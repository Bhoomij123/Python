```python
import pandas as pd

df = pd.read_csv("bike.csv")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head(10)
```

```python
df.isnull().sum()
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt

plt.hist(df['selling_price'])
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.title("Distribution of Selling Price")
plt.show()
```

```python
avg_price = df.groupby('seller_type')['selling_price'].mean()
avg_price.plot(kind='bar')
plt.title("Average Selling Price by Seller Type")
plt.ylabel("Average Price")
plt.show()
```

```python
avg_km = df.groupby('owner')['km_driven'].mean()
avg_km.plot(kind='bar')
plt.title("Average km Driven by Ownership Type")
plt.ylabel("Avg KM Driven")
plt.show()
```

```python
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

before_stats = df['km_driven'].describe()

df = df[(df['km_driven'] >= lower) & (df['km_driven'] <= upper)]

after_stats = df['km_driven'].describe()

before_stats, after_stats
```

```python
plt.scatter(df['year'], df['selling_price'])
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.title("Bike Age vs Selling Price")
plt.show()
```

```python
seller_dummies = pd.get_dummies(df['seller_type'], prefix='seller')
df_encoded = pd.concat([df, seller_dummies], axis=1)
df_encoded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df_encoded.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

#  What are the most important factors affecting a bike's selling price?
#  Mention any data cleaning or feature engineering you performed.*bold text* **

From the analysis, the most important factors affecting a bikes selling price are:

Ex-showroom price  Higher original price bikes retain more value, showing the strongest positive correlation.

Year (bike age)  Newer bikes sell for higher prices; price drops as bike gets older.

Ownership  Bikes with 1st owner have higher value than 2nd or 3rd owners.

KM Driven  More driven bikes generally have lower price.

Seller Type  Dealers sell bikes at higher prices than individuals.

 Data Cleaning & Feature Engineering Performed

Missing values handled using median for numeric & mode for categorical fields.

Removed extreme outliers in km_driven using the IQR method.

Converted seller_type into numeric form using one-hot encoding.

Corrected data types and checked for duplicates.

Created visualizations to understand trends and relationships.

 Final Insight

Newer bikes with lower mileage, fewer owners, and originally higher showroom prices have the highest selling price in the market.
