## Question 1: What is Deep Learning? Explain its evolution and differences from traditional Machine Learning

Deep Learning is a specialized subset of Machine Learning that uses **Artificial Neural Networks (ANNs)**
with multiple hidden layers to model complex patterns in data. These deep architectures allow the model
to automatically learn hierarchical feature representations from raw data.

### Evolution of Deep Learning
- **1950s–60s:** Perceptron introduced by Frank Rosenblatt  
- **1980s:** Multilayer neural networks and backpropagation  
- **1990s:** Limited progress due to lack of data and computation  
- **2010 onwards:** Breakthrough due to:
  - Availability of large datasets
  - Powerful GPUs
  - Improved activation functions and optimizers

### Difference from Traditional Machine Learning
| Aspect | Traditional ML | Deep Learning |
|------|----------------|---------------|
| Feature Engineering | Manual | Automatic |
| Data Requirement | Small–Medium | Large |
| Computation | Low | High |
| Use Cases | Structured data | Images, audio, text |



## Question 2: Perceptron Architecture and Limitations

A perceptron is the simplest neural network model used for binary classification.

### Architecture
- Input layer
- Weights
- Bias
- Summation unit
- Activation function (step function)

### Working
The perceptron computes a weighted sum of inputs and passes it through an activation function.

### Limitations
- Can only solve **linearly separable problems**
- Fails for XOR problem
- No hidden layers



## Question 3: Activation Functions – Sigmoid, ReLU, Tanh

Activation functions introduce **non-linearity** into neural networks.

### Sigmoid
- Output range: (0,1)
- Problem: vanishing gradient

### ReLU
- Output: max(0, x)
- Fast and efficient
- Problem: dying ReLU

### Tanh
- Output range: (-1,1)
- Zero-centered but computationally expensive



## Question 4: Loss Function vs Cost Function

- **Loss Function:** Error for a single training sample  
- **Cost Function:** Average loss over entire dataset  

### Examples
- Mean Squared Error (Regression)
- Binary Crossentropy (Classification)



## Question 5: Optimizers in Neural Networks

Optimizers adjust weights to minimize loss.

### Gradient Descent
- Fixed learning rate
- Slow convergence

### RMSprop
- Adaptive learning rate
- Solves vanishing gradient problem

### Adam
- Combines Momentum + RMSprop
- Fast and widely used


## Question 6: Single Layer Perceptron for AND Gate (NumPy)

```python

import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.random.rand(2)
b = 0.0
lr = 0.1

for epoch in range(100):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = 1 if z >= 0 else 0
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

print("Weights:", w)
print("Bias:", b)
```

## Question 7: Activation Function Visualization

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,200)
sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)
tanh = np.tanh(x)

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")
plt.legend()
plt.title("Activation Functions")
plt.show()
```

## Question 8: MNIST Classification using Keras

```python

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)
```

## Question 9: Fashion MNIST – Loss & Accuracy Interpretation

```python

from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
X_train = X_train / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```
